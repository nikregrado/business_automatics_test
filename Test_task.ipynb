```python
import pandas as pd
import numpy
```

```python
url = 'https://query.data.world/s/OuZzRk6mqxRar7_DS3PWH0v5Khy9Qw'
df = pd.read_csv(url)
df
```

```python
df = df.fillna('0')
df = df.astype({'pclass': 'int64', 'survived': 'int64', 'age': 'int64', 'sibsp': 'int64', 'parch': 'int64', 'body': 'int64', 'fare': 'float64'})

df['name'] = df['name'].str.split(',').str[0]

df['FamilySize'] = df['sibsp'] + df['parch'] + 1
df['type'] = 1 
df['type'].loc[df['FamilySize'] > 1] = 0
df = df.sort_values(by='name')

df = df.drop({'pclass', 'sex', 'ticket', 'fare', 'cabin', 'embarked', 'home.dest', 'body'}, axis=1)

# df.groupby(['type','name']).agg({'survived': 'sum', 'age': 'mean', 'boat': lambda x: ','.join(x), 'Family_size': 'max'})
new_df = pd.DataFrame(df.groupby(['type','name']).agg({'survived': 'sum', 'age': 'mean', 'boat': lambda x: ','.join(x), 'FamilySize': 'max'}))
new_df
```

```python
new_df.to_csv('new.csv')
```
